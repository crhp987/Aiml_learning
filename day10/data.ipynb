```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Asha", "Rahul", "Sita", None, "Kiran"],
    "Age": [23, np.nan, 29, 25, None],
    "Salary": [45000, 52000, None, 48000, 51000],
    "City": ["Bangalore", "Mysuru", "Bangalore", "Mysuru", None]
}

df = pd.DataFrame(data)
print(df)
```

```python
#missing values
df.isnull()
```

```python
#count missing values
df.isnull().sum()
```

```python
#
df.isnull().any()
```

```python

df.dropna()
```

```python

df.dropna(axis=1)
```

```python
df.fillna(0)
```

```python
df["Age"].fillna(df["Age"].mean(),  inplace=true)
```

```python
df["Salary"].fillna(df["Salary"].median(),  inplace=true)
```

```python
df.fillna(method="ffill")
df.fillna(method="bfill")
```

```python
df.replace("unknown", np.nan, inplace=True)
df.replace(-1, np.nan, inplace=True)
```
