```python
import pandas as pd

data = {
    "Age": ["23", "29", "25", "30"],
    "Salary": ["45000", "52000", "48000", "51000"],
    "Join_Date": ["2023-01-10", "2022-12-05", "2023-02-20", "2022-11-15"]
}

df = pd.DataFrame(data)
print(df.dtypes)
```

```python
#convert strint to interger
df["Age"] = df["Age"].astype(int)
print(df.dtypes)
```

```python
#string to float
df["Salary"] = df["Salary"].astype(float)
print(df.dtypes)
```

```python
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
print(df.dtypes)
```
