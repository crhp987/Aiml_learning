```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel("days")
plt.ylabel("sales")
plt.title("sales growth over days")

plt.show()
```

```python
plt.plot(x, y, color='green', linestyle='--', marker=0)
plt.show()
```

```python
y2 = [5, 15, 20, 25, 35]
plt.plot(x, y, label="product A")
plt.plot(x, y2, label="product B")

plt.xlabel("days")
plt.ylabel("sales")
plt.title("sales growth over days")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
hours = [1, 2, 3, 4, 5]
marks = [50, 60, 38, 45, 60]

plt.scatter(hours, marks)
plt.xlabel("hours")
plt.ylabel("marks")
plt.title("outlier detection ")
plt.show()
```

```python
import matplotlib.pyplot as plt


subjects = ["Math", "Science", "English", "History"]
scores = [85, 90, 78, 88]


plt.bar(subjects, scores)


plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")

plt.show()
```
