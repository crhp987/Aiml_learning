```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y = [2, 4, 6, 8, 10]
categories = ["A", "B", "C", "D", "E"]
values = [5, 7, 3, 8, 6]
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
hist_data = np.random.randn(100)

plt.figure(figsize=(10, 8))

# Line chart
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("Line Chart")

# Bar chart
plt.subplot(2, 2, 2)
plt.bar(categories, values)
plt.title("Bar Chart")

# Scatter plot
plt.subplot(2, 2, 3)
plt.scatter(scatter_x, scatter_y)
plt.title("Scatter Plot")

# Histogram
plt.subplot(2, 2, 4)
plt.hist(hist_data, bins=10)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```
