```python
import matplotlib.pyplot as plt
import numpy as np

def create_sales_dashboard():
    """
    Generates a dashboard with a bar chart and a line plot using subplots.
    """
    
    # Create a figure and a set of subplots
    plt.figure(figsize=(10, 5))

    #Categorical Sales Comparison (Bar Chart) ---
    plt.subplot(1, 2, 1)
    categories = ['Electronics', 'Clothing', 'Home']
    sales_values = [300, 450, 200]
    
    plt.bar(categories, sales_values, color=['blue', 'orange', 'green'])
    plt.title('Sales per Product Category')
    plt.xlabel('Category')
    plt.ylabel('Sales (Units)')

    #Sales Trend Line Plot (Line Chart) ---
    plt.subplot(1, 2, 2)
    months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
    trend_values = np.array([50, 60, 55, 70, 85, 90])
    
    plt.plot(months, trend_values, marker='o', linestyle='-', color='red')
    plt.title('Monthly Sales Trend')
    plt.xlabel('Month')
    plt.ylabel('Sales (Units)')
    plt.grid(True)

    plt.tight_layout()

    # Display the dashboard
    plt.show()

if __name__ == "__main__":

    create_sales_dashboard()
```
