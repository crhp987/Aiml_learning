```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = {
    "Price": [120000, 150000, 180000, 220000, 300000, 500000, 750000],
    "City": ["NY", "LA", "NY", "LA", "Chicago", "NY", "Chicago"]
}
df = pd.DataFrame(data)

# Histogram with KDE
sns.histplot(df["Price"], kde=True)
plt.title("Price Distribution")
plt.show()

# Skewness and Kurtosis
skewness = df["Price"].skew()
kurtosis = df["Price"].kurt()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)

# Count plot for categorical column
sns.countplot(x="City", data=df)
plt.title("City Frequency")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV file
df = pd.read_csv("housing.csv")   # change filename if needed

# Histogram with KDE for numerical column
sns.histplot(df["Price"], kde=True)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.show()

# Skewness and Kurtosis
print("Skewness:", df["Price"].skew())
print("Kurtosis:", df["Price"].kurt())

# Count plot for categorical column
sns.countplot(x="City", data=df)
plt.title("City Frequency")
plt.show()
```
