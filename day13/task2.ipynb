```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Price": [120000, 150000, 180000, 220000, 300000, 500000, 750000],
    "SquareFootage": [800, 1000, 1200, 1500, 2000, 3000, 4000],
    "City": ["NY", "LA", "NY", "LA", "Chicago", "NY", "Chicago"]
}
df = pd.DataFrame(data)

# Scatter Plot: SquareFootage vs Price
sns.scatterplot(x="SquareFootage", y="Price", data=df)
plt.title("House Size vs Price")
plt.xlabel("Square Footage")
plt.ylabel("Price")
plt.show()

# Boxplot: City vs Price
sns.boxplot(x="City", y="Price", data=df)
plt.title("Price by City")
plt.show()

# Observation
print("Observation: As SquareFootage increases, Price seems to increase.")
```
