```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Price": [120000, 150000, 180000, 220000, 300000, 500000, 750000],
    "SquareFootage": [800, 1000, 1200, 1500, 2000, 3000, 4000],
    "Bedrooms": [2, 3, 3, 4, 4, 5, 6],
    "Bathrooms": [1, 2, 2, 3, 3, 4, 4]
}

df = pd.DataFrame(data)

#Correlation matrix
corr_matrix = df.corr()
print("Correlation Matrix:\n", corr_matrix)

#Visualize correlation matrix with heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Identify pairs with correlation > 0.8
high_corr = [(i, j, corr_matrix.loc[i,j]) for i in corr_matrix.columns for j in corr_matrix.columns 
             if i != j and abs(corr_matrix.loc[i,j]) > 0.8]
print("Highly correlated pairs (>0.8):", high_corr)

# Boxplot to detect outliers in main numerical column (Price)
sns.boxplot(y=df["Price"])
plt.title("Boxplot of Price")
plt.show()
```
