```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = pd.DataFrame({
    "Age": [22, 25, 30, 35, 40, 45, 50],
    "Salary": [20000, 30000, 50000, 70000, 90000, 120000, 150000]
})
```

```python
std_scaler = StandardScaler()
standardized = std_scaler.fit_transform(data)
standardized_df = pd.DataFrame(standardized, columns=data.columns)
```

```python
minmax_scaler = MinMaxScaler()
normalized = minmax_scaler.fit_transform(data)
normalized_df = pd.DataFrame(normalized, columns=data.columns)
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(data["Salary"])
plt.title("Original Salary")

plt.subplot(1, 3, 2)
plt.hist(standardized_df["Salary"])
plt.title("Standardized Salary")

plt.subplot(1, 3, 3)
plt.hist(normalized_df["Salary"])
plt.title("Normalized Salary")

plt.tight_layout()
plt.show()
```

```python
print("Original Data:\n", data)
print("\nStandardized Data:\n", standardized_df)
print("\nNormalized Data:\n", normalized_df)
```
