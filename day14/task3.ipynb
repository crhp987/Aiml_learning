```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

#Create curved (non-linear) data
np.random.seed(0)
X = np.linspace(-3, 3, 100).reshape(-1, 1)

# True curved relationship: y = x^2 + noise
y = X**2 + np.random.normal(0, 1, 100).reshape(-1, 1)

#Train normal Linear Regression
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)

r2_linear = r2_score(y, y_pred_linear)

#Create Polynomial Features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)

r2_poly = r2_score(y, y_pred_poly)

#Print R² scores
print("R² score (Linear):", r2_linear)
print("R² score (Polynomial):", r2_poly)

#Plot results
plt.scatter(X, y, label="Data")
plt.plot(X, y_pred_linear, label="Linear Fit")
plt.plot(X, y_pred_poly, label="Polynomial Fit")
plt.legend()
plt.title("Linear vs Polynomial Regression")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Curved data
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = X**2 + np.random.normal(0, 1, 100).reshape(-1, 1)

# Linear model
lin = LinearRegression().fit(X, y)
r2_lin = r2_score(y, lin.predict(X))

# Polynomial model
poly = PolynomialFeatures(2)
X_poly = poly.fit_transform(X)
poly_model = LinearRegression().fit(X_poly, y)
r2_poly = r2_score(y, poly_model.predict(X_poly))

print("R² Linear:", r2_lin)
print("R² Polynomial:", r2_poly)
```
