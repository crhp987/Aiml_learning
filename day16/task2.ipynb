```python
import numpy as np
import pandas as pd

# Sample dataset (example: student scores)
np.random.seed(42)
data = np.random.normal(loc=70, scale=10, size=100)

# Add some extreme values (outliers)
data = np.append(data, [120, 130])

df = pd.DataFrame({"Scores": data})

#Calculate Mean and Standard Deviation
mean = df["Scores"].mean()
std = df["Scores"].std()

#Calculate Z-score
df["z_score"] = (df["Scores"] - mean) / std

#Identify outliers (|Z| > 3)
outliers = df[np.abs(df["z_score"]) > 3]

print("Mean:", mean)
print("Standard Deviation:", std)
print("\nOutliers (|Z| > 3):")
print(outliers)
```
