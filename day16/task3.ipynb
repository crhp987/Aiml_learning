```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For reproducibility
np.random.seed(42)

#Create a heavily right-skewed dataset (like income)
original_data = np.random.exponential(scale=2, size=10000)

#Take 1000 samples of size 30 and store their means
sample_means = []

for _ in range(1000):
    sample = np.random.choice(original_data, size=30)
    sample_means.append(np.mean(sample))

#Plot original data distribution
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(original_data, kde=True)
plt.title("Original Data (Right-Skewed)")

# Step 4: Plot distribution of sample means
plt.subplot(1,2,2)
sns.histplot(sample_means, kde=True)
plt.title("Distribution of Sample Means (Bell Curve)")

plt.tight_layout()
plt.show()
```
