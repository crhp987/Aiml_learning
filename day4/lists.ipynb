```python
contacts = {
    "poornima": "345-1234",
    "ram": "555-2345",
    "shyam": "555-3456"
}

contacts["poornima"] = "345-4567"   
contacts["shyam"] = "555-9999"     

existing_lookup = contacts.get("poornima", "Contact not found")
missing_lookup = contacts.get("ananya", "Contact not found")

for name, phone in contacts.items():
    print(f"Contact: {name} | Phone: {phone}")

print("\nLookup existing:", existing_lookup)
print("Lookup missing:", missing_lookup)
```
