```python
import csv

data = [
    ["Name", "Grade", "Status"],
    ["Alice", "A", "Pass"],
    ["Bob", "B", "Pass"],
    ["Charlie", "F", "Fail"]
]

with open('students.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("students.csv file created.")

```
