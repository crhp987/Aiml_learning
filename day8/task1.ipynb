```python
import numpy as np

#Create a 5x3 array of random scores (50â€“100)
scores = np.random.randint(50, 101, size=(5, 3))

#Column-wise mean (mean score per subject)
mean_scores = scores.mean(axis=0)

#Subtract mean using broadcasting
centered_scores = scores - mean_scores

# Output
print("Original scores:\n", scores)
print("\nMean per subject:\n", mean_scores)
print("\nCentered scores:\n", centered_scores)
```
