```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva" ,"ram"],
    "Math": [78, 45, 88, 60, 35, 80],
    "Science": [85, 50, 92, 58, 40, 50],
    "English": [90, 40, 84, 65, 30, 45],
    "Attendance": [92, 80, 95, 85, 70, 65]
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)

print("students.csv created successfully!")
```

```python
import pandas as pd
import numpy as np

# Read CSV file
df = pd.read_csv("students.csv")

print("Student Data:\n", df)

# Average marks per student
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)


# Pass/Fail classification
def classify(avg):
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

df["Result"] = df["Average"].apply(classify)

# NumPy analysis
marks_array = df[["Math", "Science", "English"]].to_numpy()

print("\nNumPy Analysis:")
print("Overall Average:", np.mean(marks_array))
print("Maximum Marks:", np.max(marks_array))
print("Minimum Marks:", np.min(marks_array))

# Filtering pass/fail students
passed_students = df[df["Result"] == "Pass"]
failed_students = df[df["Result"] == "Fail"]

print("\nPassed Students:\n", passed_students)
print("\nFailed Students:\n", failed_students)


# Subject-wise topper
subjects = ["Math", "Science", "English"]

print("\nSubject-wise Toppers:")
for subject in subjects:
    topper = df.loc[df[subject].idxmax()]
    print(f"{subject} Topper: {topper['Name']} ({topper[subject]})")

# Attendance vs marks analysis
correlation = df["Attendance"].corr(df["Average"])

print("\nAttendance vs Average Marks Correlation:", correlation)

# Pandas Series example
avg_series = pd.Series(df["Average"], name="Average Marks")
print("\nPandas Series (Average Marks):\n", avg_series)

# Final DataFrame
print("\nFinal Data with Derived Columns:\n", df)
```
