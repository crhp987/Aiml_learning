```python
import pandas as pd

# Student attendance data
names = ["Amith", "Sara", "Sony"]
days_present = [22, 25, 18]
total_days = [25, 25, 25]

# Create DataFrame
data = {
    "Name": names,
    "Days_Present": days_present,
    "Total_Days": total_days
}

df = pd.DataFrame(data)

df["Days_Absent"] = df["Total_Days"] - df["Days_Present"]            
df["Attendance_%"] = (df["Days_Present"] / df["Total_Days"]) * 100  
df["Bonus_Days"] = df["Days_Present"] + 2                            
df["Double_Present"] = df["Days_Present"] * 2                         


mean_attendance = df["Days_Present"].mean()      
max_attendance = df["Days_Present"].max()      
min_attendance = df["Days_Present"].min()       
std_attendance = df["Days_Present"].std()       

# Display DataFrame
print(df)

# Display statistics
print("\nStatistics:")
print("Mean Attendance:", mean_attendance)
print("Max Attendance:", max_attendance)
print("Min Attendance:", min_attendance)
print("Std Deviation:", std_attendance)
```
