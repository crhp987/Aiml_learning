```python
import pandas as pd

# 1. Create the Series with custom labels
data = [700, 150, 300, 50]
index = ['Laptop', 'Mouse', 'Keyboard', 'Cable']
products = pd.Series(data, index=index)

print("--- Full Product Series ---")
print(products)

# 2. Access 'Laptop' price using label-based indexing
laptop_price = products['Laptop']
print(f"\nPrice of Laptop: {laptop_price}")

# 3. Slice first two products using positional indexing (0-based)
first_two = products[0:2]
print("\nFirst Two Products:")
print(first_two)
```
