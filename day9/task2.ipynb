```python
import pandas as pd
import numpy as np

grades = pd.Series([85, None, 92, 45, None, 78, 55])
print("--- Original Series ---")
print(grades)

#Identify missing values using .isnull()
missing_mask = grades.isnull()
print("\n--- Missing Value Mask ---")
print(missing_mask)

#Fill missing values with a default score of 0
filled_grades = grades.fillna(0)
print("\n--- Filled Series (Missing = 0) ---")
print(filled_grades)

high_scores = filled_grades[filled_grades > 60]
print("\n--- Filtered Results (Scores > 60) ---")
print(high_scores)
```
