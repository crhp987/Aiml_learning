```python
import pandas as pd

usernames = pd.Series([' Alice ', 'bOB', ' Charlie_Data ', 'daisy'])
usernames = usernames.str.strip()

usernames = usernames.str.lower()

contains_a = usernames.str.contains('a')

print("Processed Usernames:")
print(usernames)

print("\nContains letter 'a'?")
print(contains_a)
```
